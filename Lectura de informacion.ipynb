```python
import matplotlib.pyplot as plt
```

```python
def graficar(titulo):
    plt.ylabel("Tiempo [s]")
    plt.xlabel("Cantidad de Datos")
    plt.title(titulo)
    plt.show()

def graficar_lineal(x, y, color, titulo):
    plt.plot(x, y, color)
    plt.legend(['Linear Search'], prop = {'size':10}, loc = 'upper left')
    graficar(titulo)

def graficar_binaria(x, y, titulo, *datos):
    colores = ("r", "g", "b")
    i = 0
    for p in y:
        n = i % len(colores)
        plt.plot(x, p, colores[n])
        i += 1
    plt.legend(("paralelo", "serial"), prop = {'size':10}, loc = 'upper left')
    graficar(titulo)
```

```python
def obtener_info(nombre_archivo):
    tiempo = []
    cantidad = []
    archivo = open(nombre_archivo, "r")
    for linea in archivo.readlines():
        linea = linea.split(",")
        cantidad.append(int(linea[0]))
        tiempo.append(float(linea[1]))
    archivo.close()
    return cantidad, tiempo
```

```python
def obtener_info_archivos():
    carpeta_padre = "10000-20000/"  # Respaldo/ รณ datos/
    casos = ("paralelo", "serial")
    color = ("r","g","b")
    nombres_archivos = ("/peor_caso.txt", "/caso_promedio.txt", "/mejor_caso.txt")
    for nombre_archivo in nombres_archivos:
        titulo = nombre_archivo[1:-4]
        x1, y1 = obtener_info(carpeta_padre + casos[0] + nombre_archivo)
        x2, y2 = obtener_info(carpeta_padre + casos[1] + nombre_archivo)
        graficar_binaria(x1, [y1, y2], titulo, "paralelo", "serial")
```

```python
obtener_info_archivos()
```
